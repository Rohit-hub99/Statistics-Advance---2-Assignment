# 1. What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or inferences about a population based on a sample. It involves testing an assumption (hypothesis) about a population parameter using sample data.


# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis is the default assumption that there is no effect, no difference, or no relationship in the population.

  - It represents the status quo or what is assumed to be true unless there is strong evidence against it.

      - Alternative Hypothesis (H₁ or Ha):
        - The alternative hypothesis is what we are trying to prove—it suggests there is an effect, a difference, or a relationship.
        - It contradicts H₀ and represents the research claim.

  - Difference between null hypothesis and alternative hypotheesis.
    
    - Null Hypothesis (H₀)
      - Meaning: No effect, no difference, no relationship
      - Assumption: Considered true unless evidence suggests otherwise
      - Goal of Testing: Try to reject H₀ if evidence is strong
      - Example Claim: "No difference between two groups"
      - Decision: If data supports H₁, reject H₀

    - Alternative Hypothesis (H₁):
      - Meaning: Effect, difference, or relationship exists
      - Assumption: Proposed by researchers
      - Goal of Testing: Try to find evidence supporting H₁
      - Example Claim: "A difference exists between two groups"
      - Decision: If data doesn’t support H₁, fail to reject H₀


# 3. What is the significance level in hypothesis testing, and why is it important?
  - 1. What is the Significance Level (α):
    -  The significance level (denoted as α) is the probability of rejecting the null hypothesis (H₀) when it is actually true (Type I error). It defines the threshold for statistical significance and is chosen before conducting the test.

    - 2. Common Values of α:
      - 0.05 (5%) → Most common choice
      - 0.01 (1%) → Used when stronger evidence is needed
      - 0.10 (10%) → Used in some exploratory studies

   - Why is the Significance Level Important:-
     - Controls Type I Error (False Positive) → Limits the chance of concluding an effect exists when it doesn’t.
     - Determines p-value Interpretation → If p ≤ α, reject H₀ (significant result); if p > α, fail to reject H₀ (not significant).
     - Ensures Statistical Rigor → Helps maintain credibility in research conclusions.


# 4. What does a P-value represent in hypothesis testing ?
  - The p-value (probability value) measures the strength of evidence against the null hypothesis (H₀). It represents the probability of obtaining the observed results (or more extreme) if H₀ is true.

  - Why is the P-value Important:-
    - Helps in making data-driven decisions.
    - Quantifies the strength of evidence against H₀.
    - Avoids subjective judgment in hypothesis testing.


# 5.  How do you interpret the P-value in hypothesis testing ?
  - The p-value is a measure of the probability of obtaining the observed results (or more extreme) assuming the null hypothesis (H₀) is true. It helps us decide whether to reject or fail to reject H₀.

  - Example Interpretation:
    - A company tests if a new marketing strategy increases sales.
      - H₀: "The new strategy does not increase sales."
      - H₁: "The new strategy increases sales."
      - After analysis, p = 0.02, and α = 0.05.

    - Since p < α (0.02 < 0.05), we reject H₀ and conclude that the new strategy significantly increases sales.


# 6. What are Type 1 and Type 2 errors in hypothesis testing?
  - When conducting hypothesis testing, we make decisions based on sample data, but there's always a risk of making errors. These errors are classified as Type I and Type II errors.

  - Type I Error (False Positive):
    - Occurs when we wrongly reject the null hypothesis (H₀) when it is actually true.
    - This means we detect an effect that doesn’t actually exist.
    - The probability of making a Type I error is represented by α (significance level).

  - 2. Type II Error (False Negative):
    - Occurs when we fail to reject the null hypothesis (H₀) when it is actually false.
    - This means we miss detecting a real effect.
    - The probability of making a Type II error is β (beta).
    - The power of a test (1 - β) is the probability of correctly rejecting H₀ when it is false.



# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
  - Key Differences Between One-Tailed and Two-Tailed Tests:
    
    - 1. One-Tailed Test:
       - Tests for: Effect in one direction
       - Alternative Hypothesis (H₁): μ > μ₀ or μ < μ₀
       - Significance Level (α): Entire α in one tail
       - P-value Calculation: Probability in one tail
       - More Powerful: More power to detect a directional effect

    - 2. Two-Tailed Test:
       - Tests for: Effect in either direction
       - Alternative Hypothesis (H₁): μ ≠ μ₀
       - Significance Level (α): α split between two tails (α/2 in each)
       - P-value Calculation: Probability in both tails
       - More Powerful: Less power (due to splitting α)


# 8. What is the Z-test, and when is it used in hypothesis testing ?
  - A Z-test is a statistical test used to determine whether there is a significant difference between a sample statistic (such as the mean) and a population parameter, assuming that the population is normally distributed and the sample size is large (n ≥ 30).
    

    - When is the Z-test Used:
      - The sample size is large (n ≥ 30) OR the population standard deviation (σ) is known.
      - The data is approximately normally distributed (especially for small samples).
      - You are testing the mean or proportions of a population.


# 9. How do you calculate the Z-score, and what does it represent in hypothesis testing ?
  - A Z-score (also called a standard score) measures how many standard deviations a data point or sample mean is from the population mean.

  - In hypothesis testing, the Z-score helps determine whether to reject the null hypothesis (H₀) by comparing it to a critical value from the Z-table.

  - Formula for Z-Score:
    
    -  Z = observed value - population mean / population standard deviation



# 10. What is the T-distribution, and when should it be used instead of the normal distribution ?
  - The T-distribution (Student's t-distribution) is a probability distribution that is similar to the normal distribution, but has heavier tails. It is used when the sample size is small (n < 30) and/or the population standard deviation (σ) is unknown.

  - Real-World Example:
     - A factory claims the average weight of a product is 500g.
       - A small sample (n = 12) is tested, yielding a mean of 495g with s = 10g.
       - Since σ is unknown and n < 30, we use a T-test instead of a Z-test.



# 11. What is the difference between a Z-test and a T-test ?
  - Z-Test vs. T-Test: Key Differences

   - 1. 	Z-Test:
      - Used When: Sample size is large (n ≥ 30) OR population standard deviation (σ) is known
      - Distribution Used: Normal distribution (Z-distribution)
      - Standard Deviation Used: Population standard deviation (σ)
      - Shape of Distribution: Fixed shape (standard normal)
      - Degrees of Freedom (df): Not needed
      - When Sample Size Increases: Always follows normal distribution
      - Use Case: Large datasets, known σ, comparing means or proportions

   - 2. T-Test:
      - Used When: Sample size is small (n < 30) OR population standard deviation (σ) is unknown
      - Distribution Used: Student's T-distribution
      - Standard Deviation Used: Sample standard deviation (s)
      - Shape of Distribution	Shape of Distribution: Changes with sample size (heavier tails for small n)
      - Degrees of Freedom (df): df=n−1
      - When Sample Size Increases: Approaches normal distribution as n increases
      - Use Case: Small datasets, unknown σ, comparing means



# 12. What is the T-test, and how is it used in hypothesis testing ?
  - The T-test is a statistical test used to determine whether there is a significant difference between the means of one or two groups when the sample size is small (n < 30) and the population standard deviation (σ) is unknown.

  - It helps test the null hypothesis (H₀) that there is no significant difference between the compared means.

    - Example Scenarios:
      - Example: A company claims that employees work an average of 40 hours per week. A sample of 10 employees shows an average of 38 hours with a standard deviation of 3 hours.

      - Question: Is the work time significantly different?
      - Use a One-Sample T-Test to compare the sample mean (38 hours) with the population mean (40 hours).

     - Example: A researcher tests whether men and women have different average exam scores.
      - Use an Independent T-Test to compare the means of two separate groups.



# 13. What is the relationship between Z-test and T-test in hypothesis testing ?
  - Both Z-test and T-test are used to test hypotheses about population means, but they differ in the conditions under which they are applied. The T-test generalizes the Z-test for small samples and unknown population standard deviations.

  - Similarities Between Z-Test and T-Test:
    - Both tests compare a sample mean to a population mean (or compare two sample means).
    - Both use a test statistic to determine statistical significance.
    - Both assume the data is normally distributed or that the Central Limit Theorem (CLT) applies.
    - Both follow the general formula:
       - Test Statistic = Observed Difference/ Standard Error



# 14. What is a confidence interval, and how is it used to interpret statistical results ?
  - A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a specified level of confidence.

  - Interpreting a Confidence Interval:
    - If a 95% confidence interval for the mean weight of a sample of fish is (250g, 300g), it means:
     - We are 95% confident that the true average weight of all fish in the population lies between 250g and 300g.
     - It does not mean there is a 95% probability that the population mean is within this range—either it is or it isn’t. The confidence level refers to the long-run frequency of such intervals capturing the true parameter.
  
  - Uses of Confidence Intervals in Statistics:
    - Hypothesis Testing: If a confidence interval does not contain the null hypothesis value (e.g., a population mean of 280g when testing if the mean is 275g), we may reject the null hypothesis.
    - Comparing Groups: If the CIs of two groups (e.g., test vs. control) do not overlap much, it suggests a significant difference between them.
    - Predicting Uncertainty: Provides a more informative summary than a single estimate by indicating the precision of the estimate.



# 15. What is the margin of error, and how does it affect the confidence interval ?
  - The margin of error (MOE) is the range within which the true population parameter is expected to fall, based on a given confidence level. It represents the maximum expected difference between the sample statistic (e.g., sample mean) and the true population parameter.

  - How MOE Affects the Confidence Interval:
    - A larger MOE results in a wider confidence interval, indicating less precision.
    - A smaller MOE results in a narrower confidence interval, indicating higher precision.

   Factors Affecting MOE:
     - Confidence Level (Z):
        - Higher confidence levels (e.g., 99%) lead to larger Z-values, increasing MOE and making the CI wider.
        - Lower confidence levels (e.g., 90%) reduce Z, making the CI narrower.
     - Sample Size (n):
        - Larger sample sizes reduce MOE, leading to a more precise estimate.
        - Smaller sample sizes increase MOE, making the CI wider.
     - Standard Deviation (σ):
        - More variability in the population (higher σ) increases MOE and widens the CI.
        - Less variability (lower σ) reduces MOE and narrows the CI.

# 19. What is the F-distribution, and when is it used in hypothesis testing ?
  - The F-distribution is a probability distribution that arises frequently in hypothesis testing, especially in comparing variances between two populations. It is positively skewed and defined by two degrees of freedom:
    - df1 (numerator degrees of freedom)
    - df2 (denominator degrees of freedom)

  - When is the F-Distribution Used?
    - 1. ANOVA (Analysis of Variance):
      - Purpose: Determines whether the means of multiple groups are significantly different.
      - Example: A researcher tests whether three different diets lead to different weight loss results.
        - F-statistic formula in ANOVA:
          - F = Between-group variance / Within-group variance
          - If the F-statistic is large, it suggests that at least one group mean is significantly different.

    - 2. Testing Equality of Variances (F-Test)
      - Purpose: Compares variances of two populations.
      - Example: A quality control team checks if two machines produce products with equal variability.
   
    - 3. Regression Analysis:
      - Purpose: Tests the overall significance of a regression model.
      - Example: In multiple linear regression, an F-test determines whether the model significantly explains the variability in the dependent variable.
        - F-statistic formula in regression:
          - F = Unexplained variance / Unexplained variance ​
          - A large F-value suggests that at least one predictor variable significantly contributes to the model.



# 20. What is an ANOVA test, and what are its assumptions ?
  - ANOVA (Analysis of Variance) is a statistical test used to compare the means of two or more groups to determine if they are significantly different. It helps analyze whether variations in the data are due to random chance or real differences among groups.
  
  - Assumptions of ANOVA:
    - For ANOVA results to be valid, the following assumptions must hold:
      - Independence:
       - Observations should be independent within and between groups.
       - Example: Students' scores in one class should not influence those in another.
      
      - Normality:
       - The dependent variable should be normally distributed within each group.
       - Can be checked using a Shapiro-Wilk test or Q-Q plot.
       - If violated, a non-parametric test (Kruskal-Wallis test) can be used instead.

      - Homogeneity of Variance (Homoscedasticity):
       - Variances across groups should be roughly equal.
       - Can be checked using Levene’s test or Bartlett’s test.
       - If violated, use Welch’s ANOVA.



# 21. What are the different types of ANOVA tests ?
  - Analysis of Variance (ANOVA) is used to compare the means of multiple groups to determine if they are significantly different. There are several types of ANOVA tests, depending on the number of independent variables (factors) and the type of data.

  - 1. One-Way ANOVA:
    - Purpose: Compares the means of one independent variable (factor) across two or more groups.
    -  Example: Testing if three different fertilizers lead to different plant growth rates.
    - Used when: There is a single categorical independent variable and one continuous dependent variable.
    - Not suitable for: Analyzing interactions between multiple factors.

  - 2. Two-Way ANOVA:
    - Purpose: Compares the means of two independent variables (factors) across multiple groups, allowing for interaction effects.
    - Example: Testing how diet type (Factor 1) and exercise level (Factor 2) affect weight loss.
    - Used when: You have two categorical independent variables and one continuous dependent variable.\
    - Not suitable for: Experiments with repeated measures (same subjects tested multiple times).
  
  - 3. Repeated Measures ANOVA:
    - Purpose: Compares means when the same subjects are tested multiple times under different conditions.
    - Example: Measuring blood pressure before, during, and after taking a medication.
    - Used when: The same subjects undergo different treatments or conditions.
    - Not suitable for: Independent groups (use One-Way or Two-Way ANOVA instead).

  - 4. Mixed-Design ANOVA (Split-Plot ANOVA):
    -  Purpose: Combines Repeated Measures ANOVA and Two-Way ANOVA, where one factor is within-subjects (repeated measures) and the other is between-subjects (independent groups).
    - Example: Measuring how different teaching methods (Factor 1, between-subjects) affect students' test scores over time (Factor 2, within-subjects).
    - Used when: You have both independent groups and repeated measurements.
    - Not suitable for: Simple independent group comparisons (use One-Way or Two-Way ANOVA).

  - 5. MANOVA (Multivariate Analysis of Variance):
    - Purpose: Extends ANOVA by analyzing multiple dependent variables at the same time.
    - Example: Studying how different exercise programs affect both weight loss and muscle gain simultaneously.
    - Used when: You have multiple dependent variables that may be correlated.
    - Not suitable for: Single dependent variable analysis (use standard ANOVA).

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
import scipy.stats as stats

# Given data
population_mean = 100  # Known population mean (μ)
population_std = 15     # Known population standard deviation (σ)
sample_data = [102, 98, 105, 110, 95, 99, 103, 107, 108, 97]

# Compute sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Multiply by 2 for two-tailed test

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("Reject H0: The sample mean is significantly different from the population mean.")
else:
    print("Fail to Reject H0: No significant difference between the sample and population mean.")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Given data
population_mean = 100  # Known population mean (μ)
population_std = 15     # Known population standard deviation (σ)
sample_size = 30        # Sample size

# Generate a random sample from a normal distribution
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Compute sample mean
sample_mean = np.mean(sample_data)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Multiply by 2 for two-tailed test

# Define significance level
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)  # Critical value for two-tailed test

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"Z-Critical (±): ±{z_critical:.2f}")

# Decision rule
if abs(z_score) > z_critical:
    print("Reject H0: The sample mean is significantly different from the population mean.")
else:
    print("Fail to Reject H0: No significant difference between the sample and population mean.")

# Visualization
x = np.linspace(-4, 4, 1000)  # Range for Z-distribution
y = stats.norm.pdf(x, 0, 1)   # Standard normal distribution

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

# Shade critical regions
plt.fill_between(x, y, where=(x < -z_critical), color='red', alpha=0.3, label="Rejection Region")
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3)

# Plot Z-score
plt.axvline(z_score, color='black', linestyle='dashed', label=f"Z-Score = {z_score:.2f}")
plt.axvline(-z_critical, color='red', linestyle='dashed', label=f"Critical Value = ±{z_critical:.2f}")
plt.axvline(z_critical, color='red', linestyle='dashed')

# Labels and legend
plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()

```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, beta=0.2, pop_mean=100, sample_mean_alt=105, pop_std=15, sample_size=30):
    """
    Visualizes Type 1 and Type 2 Errors in Hypothesis Testing.

    Parameters:
        alpha (float): Significance level (Type 1 error probability).
        beta (float): Type 2 error probability.
        pop_mean (float): Null hypothesis mean (H0).
        sample_mean_alt (float): Alternative hypothesis mean (H1).
        pop_std (float): Population standard deviation.
        sample_size (int): Sample size.
    """

    # Compute standard error
    standard_error = pop_std / np.sqrt(sample_size)

    # Critical value (Z-critical for one-tailed test)
    z_critical = stats.norm.ppf(1 - alpha)

    # Compute critical value in terms of the population mean
    critical_value = pop_mean + z_critical * standard_error

    # Define the x-axis range
    x = np.linspace(pop_mean - 4*pop_std, sample_mean_alt + 4*pop_std, 1000)

    # Compute probability density functions
    pdf_h0 = stats.norm.pdf(x, pop_mean, standard_error)  # Under H0
    pdf_h1 = stats.norm.pdf(x, sample_mean_alt, standard_error)  # Under H1

    # Plot normal distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf_h0, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, pdf_h1, label="Alternative Hypothesis (H1)", color="green")

    # Shade Type 1 Error (α)
    plt.fill_between(x, pdf_h0, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 Error (β)
    beta_value = stats.norm.cdf(critical_value, sample_mean_alt, standard_error)
    plt.fill_between(x, pdf_h1, where=(x <= critical_value), color='orange', alpha=0.3, label="Type 2 Error (β)")

    # Mark critical value
    plt.axvline(critical_value, color='black', linestyle="dashed", label=f"Critical Value = {critical_value:.2f}")

    # Labels and legend
    plt.title("Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic (Sample Mean)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()

    # Show plot
    plt.show()

    # Print numerical results
    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type 1 Error (α): {alpha}")
    print(f"Type 2 Error (β): {beta_value:.4f}")

# Call function with default parameters
visualize_errors()
```

```python
# 6.  Write a Python program to perform an independent T-test and interpret the results@

import numpy as np
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data for two independent samples
group1 = np.random.normal(loc=100, scale=15, size=30)  # Sample 1 (mean=100, std=15, n=30)
group2 = np.random.normal(loc=110, scale=15, size=30)  # Sample 2 (mean=110, std=15, n=30)

# Perform Independent (Two-Sample) T-Test (Welch's T-Test for unequal variances)
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

# Output results
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("Reject H0: The means of the two groups are significantly different.")
else:
    print("Fail to Reject H0: No significant difference between the two groups.")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95, pop_std=None):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample: List or NumPy array of sample data
    - confidence: Confidence level (default is 95%)
    - pop_std: Population standard deviation (if known)

    Returns:
    - Tuple containing (Lower Bound, Upper Bound)
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation

    # Determine critical value (Z-score for known pop_std, T-score otherwise)
    if pop_std:
        # Z-Score for given confidence level
        z_critical = stats.norm.ppf((1 + confidence) / 2)
        margin_of_error = z_critical * (pop_std / np.sqrt(sample_size))
    else:
        # T-Score for given confidence level and sample size
        t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
        margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample

# Calculate 95% CI
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: {ci}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample dataD

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95, pop_std=None):
    """
    Calculate the Margin of Error (MOE) for a given confidence level.

    Parameters:
    - sample: List or NumPy array of sample data
    - confidence: Confidence level (default is 95%)
    - pop_std: Population standard deviation (if known)

    Returns:
    - Margin of Error (MOE)
    """
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation

    # Determine critical value (Z-score if pop_std is known, T-score otherwise)
    if pop_std:
        # Z-Score for given confidence level
        critical_value = stats.norm.ppf((1 + confidence) / 2)
        std_error = pop_std / np.sqrt(sample_size)
    else:
        # T-Score for given confidence level and sample size
        critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
        std_error = sample_std / np.sqrt(sample_size)

    # Compute Margin of Error
    moe = critical_value * std_error
    return moe

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample

# Calculate Margin of Error (assuming population std dev is unknown)
moe_result = margin_of_error(sample_data)
print(f"Margin of Error (95% Confidence Level): {moe_result:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, evidence):
    """
    Apply Bayes' Theorem to compute posterior probability.

    Parameters:
    - prior: P(H) - Prior probability of hypothesis
    - likelihood: P(E | H) - Likelihood of evidence given hypothesis
    - evidence: P(E) - Total probability of evidence

    Returns:
    - Posterior probability P(H | E)
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Given probabilities
P_H = 0.01   # Prior: Probability of having the disease
P_E_H = 0.95  # Likelihood: Probability of positive test given disease
P_E_not_H = 0.05  # False positive rate: Probability of positive test given no disease
P_not_H = 1 - P_H  # Probability of NOT having the disease

# Total probability of a positive test (Law of Total Probability)
P_E = (P_E_H * P_H) + (P_E_not_H * P_not_H)

# Compute Posterior: Probability of having the disease given a positive test
P_H_E = bayes_theorem(P_H, P_E_H, P_E)
print(f"Probability of having the disease given a positive test: {P_H_E:.4f}")
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples

import numpy as np
import scipy.stats as stats

# Generate random samples (normally distributed)
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15, Size=30

# Calculate sample variances
var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
var2 = np.var(sample2, ddof=1)  # Sample variance (unbiased)

# Compute F-statistic
F_statistic = var1 / var2

# Compute p-value (two-tailed test)
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

# Print results
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-Statistic: {F_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\n Reject the Null Hypothesis: Variances are significantly different.")
else:
    print("\n Fail to Reject the Null Hypothesis: No significant difference in variances.")
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats

def check_anova_assumptions(*groups):
    """
    Function to check ANOVA assumptions:
    1. Normality (Shapiro-Wilk Test)
    2. Homogeneity of Variance (Levene’s Test)

    Parameters:
        *groups: multiple groups of sample data

    Returns:
        Prints test results for normality and equal variance.
    """

    print("\n Checking ANOVA Assumptions...\n")

    # 1️⃣ Normality Test (Shapiro-Wilk)
    print(" Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"  Group {i}: W={stat:.4f}, p={p_value:.4f}  {' Normal' if p_value > 0.05 else ' Not Normal'}")

    # 2️⃣ Homogeneity of Variance Test (Levene's Test)
    print("\n Equal Variance Test (Levene’s Test):")
    stat, p_value = stats.levene(*groups)
    print(f"  Levene’s Test: W={stat:.4f}, p={p_value:.4f}  {' Equal Variances' if p_value > 0.05 else ' Unequal Variances'}")

    print("\n If assumptions hold (p > 0.05), proceed with ANOVA.")
    print(" If violated, consider transformations or non-parametric tests (e.g., Kruskal-Wallis).")

# Example Usage
np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)
group3 = np.random.normal(loc=85, scale=10, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# Step 1: Generate sample data (assumed normal distribution)
np.random.seed(42)
population_variance = 15**2  # True variance (claimed)
sample_size = 30
sample = np.random.normal(100, np.sqrt(population_variance), sample_size)

# Step 2: Calculate sample variance
sample_variance = np.var(sample, ddof=1)  # Sample variance with (n-1) denominator

# Step 3: Compute Chi-Square test statistic
chi_square_stat = (sample_size - 1) * sample_variance / population_variance

# Step 4: Compute P-value (two-tailed test)
alpha = 0.05
df = sample_size - 1  # Degrees of freedom

# Critical values for two-tailed test
chi_critical_low = stats.chi2.ppf(alpha / 2, df)
chi_critical_high = stats.chi2.ppf(1 - alpha / 2, df)

# Compute p-value
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Step 5: Print Results
print(" Hypothesis Test for Population Variance:")
print(f"  Sample Variance: {sample_variance:.4f}")
print(f"  Claimed Population Variance: {population_variance:.4f}")
print(f"  Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"  P-Value: {p_value:.4f}")

# Step 6: Interpret Results
if p_value < alpha:
    print(" Reject Null Hypothesis: The population variance is significantly different.")
else:
    print(" Fail to Reject Null Hypothesis: No significant difference in variance.")
```

```python
# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
import scipy.stats as stats

# Step 1: Define sample data (number of successes & total observations for each group)
successes_group1 = 45  # Successful outcomes in group 1
total_group1 = 100     # Total observations in group 1

successes_group2 = 30  # Successful outcomes in group 2
total_group2 = 100     # Total observations in group 2

# Step 2: Compute sample proportions
p1 = successes_group1 / total_group1  # Proportion in group 1
p2 = successes_group2 / total_group2  # Proportion in group 2

# Step 3: Compute pooled proportion
p_pooled = (successes_group1 + successes_group2) / (total_group1 + total_group2)

# Step 4: Compute standard error
se = np.sqrt(p_pooled * (1 - p_pooled) * (1/total_group1 + 1/total_group2))

# Step 5: Compute Z-score
z_score = (p1 - p2) / se

# Step 6: Compute P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 7: Print results
print(" Two-Proportion Z-Test Results:")
print(f"  Proportion in Group 1: {p1:.4f}")
print(f"  Proportion in Group 2: {p2:.4f}")
print(f"  Z-Score: {z_score:.4f}")
print(f"  P-Value: {p_value:.4f}")

# Step 8: Interpret Results
alpha = 0.05
if p_value < alpha:
    print(" Reject Null Hypothesis: The proportions are significantly different.")
else:
    print(" Fail to Reject Null Hypothesis: No significant difference in proportions.")
```

```python
# 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate random datasets
np.random.seed(42)
sample1 = np.random.normal(100, 15, 30)  # Mean=100, Std Dev=15, Size=30
sample2 = np.random.normal(100, 10, 30)  # Mean=100, Std Dev=10, Size=30

# Step 2: Compute sample variances
var1 = np.var(sample1, ddof=1)  # Variance of sample 1
var2 = np.var(sample2, ddof=1)  # Variance of sample 2

# Step 3: Compute F-statistic (larger variance on top for right-tailed test)
F_stat = var1 / var2 if var1 > var2 else var2 / var1

# Step 4: Compute P-value (two-tailed test)
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * min(stats.f.cdf(F_stat, df1, df2), 1 - stats.f.cdf(F_stat, df1, df2))

# Step 5: Print results
print(" F-Test for Variance Comparison:")
print(f"  Variance of Sample 1: {var1:.4f}")
print(f"  Variance of Sample 2: {var2:.4f}")
print(f"  F-Statistic: {F_stat:.4f}")
print(f"  P-Value: {p_value:.4f}")

# Step 6: Interpret Results
alpha = 0.05
if p_value < alpha:
    print(" Reject Null Hypothesis: Variances are significantly different.")
else:
    print(" Fail to Reject Null Hypothesis: No significant difference in variances.")

# Step 7: Visualization - Boxplots to compare variance
plt.figure(figsize=(8, 5))
plt.boxplot([sample1, sample2], labels=['Sample 1', 'Sample 2'], patch_artist=True)
plt.title("Comparison of Sample Variances")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results

import numpy as np
import scipy.stats as stats

# Observed data
observed = np.array([50, 30, 20])  # Example observed counts

# Correct expected frequencies: Scale them to match the sum of observed
expected = np.array([33.33, 33.33, 33.33])
expected = expected * (observed.sum() / expected.sum())  # Normalize expected to match observed total

# Perform Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print(" Reject Null Hypothesis: The observed distribution is significantly different from expected.")
else:
    print(" Fail to Reject Null Hypothesis: No significant difference between observed and expected distributions.")
```
